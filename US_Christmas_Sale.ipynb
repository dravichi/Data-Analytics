**Import Essential Libraries**

```python
from google.colab import files
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**Import Dataset**

```python
files.upload()
df = pd.read_csv('Week 40 - US Christmas Tree Sales - 2010 to 2016.csv')
```

**Data Overview**

**Data Preprocessing**

```python
df.drop(columns=['index'], axis=1, inplace=True)

df.rename(columns={'Average Tree Price': 'Average tree price'}, inplace=True)

df.tail()
```

**Data Visualization**

```python
df.loc[:, 'Year':'Sales'].groupby(['Year'], as_index=False).mean(numeric_only=True)
```

```python
sns.barplot(x='Year', y='Average tree price', hue='Type of tree', data=df).set_title('Fake and Real Tree Average Price Comparison by Year')
```

```python
sns.lineplot(x='Year',y='Sales', data=df).set_title('Sales by Year')
```

```python
sns.scatterplot(x='Number of trees sold', y='Sales', data=df).set_title('No. of Tree Sold and Sales Correlation')
```
